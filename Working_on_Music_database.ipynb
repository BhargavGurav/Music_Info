## Average duration of songs of each genre

```python

df2 = df1.groupby(df1["genre"]).mean()["duration_ms"]
df2 = df2.to_dict()
for i, j in df2.items():
    print(f"{i:10}\t{j:10.2f}")
```

### Maximum and Minimum length genre 

```python
max_length_in_ms = max(df2.values())
min_length_in_ms = min(df2.values())

Max_length_Genre  = [i for i in df2 if df2[i] == max_length_in_ms][0]
Min_length_Genre = [i for i in df2 if df2[i] == min_length_in_ms][0]

print("The max length music's genre is {} and its length is {} ms\n The min lenght music's genre is {} and its length is {} ms".format(Max_length_Genre, round(max_length_in_ms, 2), Min_length_Genre, round(min_length_in_ms, 2)))
```

# Genre distribution by year: Analyze the distribution of genres over the years. Count the number of songs in each genre for different years to identify trends and changes in genre popularity over time.


```python
genres = df1["genre"].unique()
print(genres)
print(len(genres))
```

```python
count_of_genres_song = df1.groupby(["genre", "year"]).size()
```

```python
for genre in genres:
    count_of_genres_song[genre].plot(label=genre)
    plt.rcParams['figure.figsize'] = [15, 7]
plt.legend()
```

# Popular tags by artist: Identify the most commonly associated tags for each artist in the dataset. This analysis can help uncover recurring themes or characteristics associated with specific artists.


```python
artists = df1["artist"].unique()
print(artists)
print(len(artists))
```

```python
artists_and_tags = {}
for artist in artists:
    tags = set()
    for i in df1[df1["artist"] == artist]["tags"]:
        for j in i.split(", "):
            tags.add(j)
    artists_and_tags[artist] = list(tags)

artists_and_tags
```
